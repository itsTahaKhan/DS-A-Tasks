```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
diamond=pd.read_csv('Diamonds.csv')

print(diamond.shape)

diamond.isnull().sum()
diamond.dropna(axis=0, inplace=True)
diamond.isnull().sum()
diamond.head()


data = pd.read_csv('Diamonds.csv')
```

Bar Chart

```python

average_price_by_cut = data.groupby('cut')['price'].mean().reset_index()

plt.figure(figsize=(8, 6))
plt.bar(average_price_by_cut['cut'], average_price_by_cut['price'], color='skyblue')
plt.xlabel('Cut Quality')
plt.ylabel('Average Price')
plt.title('Average Price by Cut Quality')
plt.show()
```

Pie chart

```python



cut_counts = data['cut'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(cut_counts, labels=cut_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Cut Quality')
plt.show()
```

Histrogram

```python

plt.figure(figsize=(8, 6))
plt.hist(data['carat'], bins=30, edgecolor='k')
plt.xlabel('Carat')
plt.ylabel('Frequency')
plt.title('Histogram of Carat Weight')
plt.show()
```

Scales and Transformations (Log Scale for "price"):

```python



plt.figure(figsize=(8, 6))
plt.plot(data['price'])
plt.yscale('log')
plt.xlabel('Index')
plt.ylabel('Log Price')
plt.title('Log Scale for Diamond Prices')
plt.show()
```

Overplotting, Transparency, and Jitter (Scatter Plot for "carat" vs. "price"):

```python

plt.figure(figsize=(8, 6))
plt.scatter(data['carat'] + np.random.normal(0, 0.03, len(data)), data['price'], alpha=0.5, marker='o')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Scatter Plot of Carat vs. Price with Jitter')
plt.show()
```

Heat Map (Correlation Heatmap):


```python
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sb.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

Boxplot (Using the "cut" column)

```python
plt.figure(figsize=(8, 6))
sb.boxplot(x='cut', y='price', data=data, palette='Set2')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.title('Boxplot of Price by Cut Quality')
plt.show()
```

Line Plot (Using the "price" column):

```python
plt.figure(figsize=(8, 6))
plt.plot(data['price'])
plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Line Plot of Diamond Prices')
plt.show()
```
